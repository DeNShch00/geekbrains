```python
""" график окружности, элипса, гиперболы """

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# Окружность
r = 10
x = np.linspace(-r, r, 1000)
y1 = []
y2 = []
for x_ in x:
    u = math.sqrt(r**2 - x_**2)
    y1.append(u)
    y2.append(-u)
 
plt.axes().set_aspect(1)
plt.grid()
plt.axis([-r-1, r+1, -r-1, r+1])
plt.xticks([i for i in np.arange(-r-1, r+1) if i%2==0])
plt.yticks([i for i in np.arange(-r-1, r+1) if i%2==0])
plt.plot(x, y1)
plt.plot(x, y2)
```

```python
# Элипс
a = 7
b = 3
x = np.linspace(-a, a, 1000)
y1 = []
y2 = []
for x_ in x:
    u = math.sqrt((1 - (x_**2)/(a**2)) * (b**2))
    y1.append(u)
    y2.append(-u)
        
plt.axes().set_aspect(1)
plt.grid()
plt.axis([-a-1, a+1, -b-1, b+1])
plt.xticks(np.arange(-a-1, a+1))
plt.yticks(np.arange(-a-1, a+1))
plt.plot(x, y1)
plt.plot(x, y2)
```

```python
# Гипербола
a = 2
b = 1

plt.axes().set_aspect(1)
plt.grid()
plt.axis([-a-10, a+10, -b-10, b+10])
plt.xticks(np.arange(-a-10, a+10))
plt.yticks(np.arange(-a-10, a+10))

def hyperbola(from_x, to_x):
    x = np.linspace(from_x, to_x, 1000)
    y1 = []
    y2 = []
    for x_ in x:
        u = math.sqrt(((x_**2)/(a**2) - 1) * (b**2))
        y1.append(u)
        y2.append(-u)

    plt.plot(x, y1)
    plt.plot(x, y2)

hyperbola(a, a+10)
hyperbola(-10-a, -a)
```
